```python
from sklearn.datasets import load_breast_cancer
```

```python
##Breast cancer data
breast_cancer_data = load_breast_cancer()
```

```python
from sklearn.model_selection import train_test_split

#separating in the training and test sets. Hold-out technique

training_data, validation_data, training_labels, validation_labels = train_test_split(breast_cancer_data.data, breast_cancer_data.target, test_size=0.2, random_state=40)
```

```python
from sklearn.neighbors import KNeighborsClassifier

#training the data using K-Nearest Neighbor

accuracies = []

for k in range(1, 101):
  classifier = KNeighborsClassifier(n_neighbors=k)
  classifier.fit(training_data, training_labels)
  accuracies.append(classifier.score(validation_data, validation_labels))
```

```python
import matplotlib.pyplot as plt

#plotting the accuracy for the already set parameters

k_list = range(1,101)

plt.plot(k_list, accuracies)

plt.xlabel('numbers of k')
plt.ylabel('accuracy')
plt.title('accuracy x numbers of k')

plt.show()
```
